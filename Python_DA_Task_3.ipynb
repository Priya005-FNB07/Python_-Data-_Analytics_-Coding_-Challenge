```python
df.dropna(subset=['Comments'], inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df['Culmen Length (mm)']=df['Culmen Length (mm)'].fillna(df['Culmen Length (mm)'].mode()[0])
```

```python
df['Culmen Depth (mm)']=df['Culmen Depth (mm)'].fillna(df['Culmen Depth (mm)'].mean())
```

```python
df['Flipper Length (mm)']=df['Flipper Length (mm)'].fillna(df['Flipper Length (mm)'].median())
```

```python
df['Body Mass (g)'] = df['Body Mass (g)'].fillna(df['Body Mass (g)'].mode()[0])
```

```python
df['Sex']=df['Sex'].fillna(df['Sex'].mode()[0])
```

```python
df['Delta 13 C (o/oo)']=df['Delta 13 C (o/oo)'].fillna(df['Delta 13 C (o/oo)'].mode()[0])
```

```python
df['Delta 15 N (o/oo)']=df['Delta 15 N (o/oo)'].fillna(df['Delta 15 N (o/oo)'].mode()[0])
```

```python
print(df.isnull().sum())
```

##Merge:

```python
cust_data=(pd.read_excel('/content/cust_id.xlsx'))
print(df)
```

```python
sales_data=(pd.read_excel('/content/sales_data.xlsx'))
print(df)
```

```python
merge_data=pd.merge(cust_data, sales_data, left_on='Cust_Id', right_on='cust_id')
print(merge_data)
```

```python
right_merge=cust_data.merge(sales_data, how='right', left_on='Cust_Id', right_on='cust_id')
print(right_merge)
```
