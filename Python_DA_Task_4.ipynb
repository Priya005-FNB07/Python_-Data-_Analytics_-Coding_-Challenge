## Matplotlib Visualization:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/penguins_cleaned_data (1).csv')
```

```python
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
print(df['Sex'].unique())
```

```python
print(df['Sex'].mode()[0])
```

```python
df['Sex'] = df['Sex'].fillna(df['Sex'].mode()[0])
```

```python
print(df.isnull().sum())
```

#Line Chart:

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Culmen Depth (mm)'],df['Culmen Length (mm)'],linestyle='-',color='g',marker='v',alpha=0.7)
plt.title('Line Chart')
plt.xlabel('Culmen Depth (mm) ')
plt.ylabel('Culmen Length (mm)')
plt.grid(True)
plt.show()
```

## Scatter Plot:

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Delta 15 N (o/oo)'],df['Delta 13 C (o/oo)'],color='blue',alpha=0.7)
plt.title('Scatter Plot')
plt.xlabel('Delta 15 N (o/oo)')
plt.ylabel('Delta 13 C (o/oo)')
plt.grid(True)
plt.show()
```

## Bar Chart:

```python
plt.figure(figsize=(7,5))
plt.bar(df['Island'],df['Body Mass (g)'],color='orange')
plt.title('Bar Chart')
plt.xlabel('Island')
plt.ylabel('Body Mass (g)')
plt.show()
```

## Pie Chart:

```python
plt.figure(figsize=(4,4))
species_counts = df['Species'].value_counts()
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', startangle=90,explode=(0.05,0,0),shadow=True)
plt.title('Distribution of Penguin Species')
plt.show()
```

## Histogram :

```python
plt.figure(figsize=(8,5))
plt.hist(df['Flipper Length (mm)'],bins=10,color='pink',edgecolor='black')
plt.title('Histogram')
plt.xlabel('Flipper Length (mm) ')
plt.show()
```

## Box Plot:

```python
plt.figure(figsize=(6,4))
df.boxplot(column='Body Mass (g)', by='Clutch Completion', grid=False, ax=plt.gca())
plt.title('Body Mass Vs Clutch Completion')
plt.xlabel('Clutch Completion')
plt.ylabel('Body Mass (g)')
plt.tight_layout()
plt.show()
```
