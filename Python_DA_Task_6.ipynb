```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop('Comments', axis=1, inplace=True)
```

```python
print(df.isnull().sum())
```

```python
print(df['Species'].unique())
```

```python
print(df['Island'].unique())
```

```python
print(df['Sex'].unique())
```

```python
df['Sex'] = df['Sex'].replace({'.': 'nan'})
```

```python
print(df['Clutch Completion'].unique())
```

```python
print(df['Stage'].unique())
```

```python
print(df['studyName'].unique())
```

```python
df['Culmen Length (mm)']=df['Culmen Length (mm)'].fillna(df['Culmen Length (mm)'].mode()[0])
```

```python
df['Culmen Depth (mm)'] = df['Culmen Depth (mm)'].fillna(df['Culmen Depth (mm)'].mean())
```

```python
df['Flipper Length (mm)']=df['Flipper Length (mm)'].fillna(df['Flipper Length (mm)'].median())
```

```python
df['Body Mass (g)'] = df['Body Mass (g)'].fillna(df['Body Mass (g)'].mode()[0])
```

```python
df['Sex']=df['Sex'].fillna(df['Sex'].mode()[0])
```

```python
df['Delta 13 C (o/oo)']=df['Delta 13 C (o/oo)'].fillna(df['Delta 13 C (o/oo)'].mode()[0])
```

```python
df['Delta 15 N (o/oo)']=df['Delta 15 N (o/oo)'].fillna(df['Delta 15 N (o/oo)'].mode()[0])
```

```python
print(df.isnull().sum())
```

```python
df.to_csv('penguins_cleaned_data.csv', index=False)
```
