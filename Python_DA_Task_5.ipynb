# Seaborn & Plotly Advanced Visualization:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/penguins_cleaned_data (1).csv')
print(df.isnull().sum())
```

```python
df['Sex'] = df['Sex'].fillna(df['Sex'].mode()[0])
print(df.isnull().sum())
```

# Heatmap:

```python
plt.figure(figsize=(8, 5))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

#Pair Plots:

```python
plt.figure(figsize=(10, 10))
sns.pairplot(df, hue='Species')
plt.suptitle('Pair Plots')
plt.show()
```

# Violin Plot:

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Island', y='Body Mass (g)', data=df,hue='Sex')
plt.title('Violin Plot')
plt.xticks(rotation=45)
plt.show()
```

# Swarm Plot:

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x='Species', y='Body Mass (g)', data=df,hue='Sex')
plt.title('Species Vs Body Mass')
plt.xticks(rotation=45)
plt.show()
```

# Density Plot:

```python
plt.figure(figsize=(8, 4))
sns.kdeplot(data=df, x='Body Mass (g)', hue='Species', fill=True)
plt.title('Density Plot')
plt.show()
```

# Bar Chart:

```python
plt.figure(figsize=(8,4))
sns.barplot(x='Island', y='Body Mass (g)', data=df,hue='Sex')
plt.title('Bar Chart')
plt.xticks(rotation=45)
plt.show()
```

# Scatter Plot:

```python
plt.figure(figsize=(8,4))
sns.scatterplot(x='Culmen Length (mm)', y='Culmen Depth (mm)', data=df,hue='Sex')
plt.legend(loc='upper right')
plt.title('Scatter Plot')
plt.show()
```

# Line Plot:

```python
plt.figure(figsize=(8,5))
sns.lineplot(x='Island', y='Body Mass (g)', data=df,hue='Sex',linestyle=':')
plt.title('Island Vs Body Mass')
plt.xticks(rotation=45)
plt.show()
```
